```python
import pandas as pd #set new options to see all columns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
bpb = pd.read_csv("bpb_data.csv", encoding='utf8') #read in csv with correct encoding
```

```python
bpb.shape
```

```python
bpb.head(5) #show first 5 rows
```

```python
bpb = bpb.drop(bpb.index[0]) #delete first row
```

```python
bpb.head(1) #check wheter first row was deleted 
```

```python
cols = ['H2-1',  'H2-2', 'Address', 'Meta Description 1', 'Meta Keyword 1'] #bring columns into correct order
bpb = bpb[cols] #apply on dataframe
bpb.head(1) #check new column order
```

```python
bpb.columns = ['Titel', 'Untertitel', 'Link', 'Beschreibung', 'Stichworte'] #rename columns so they make more sense
bpb.head(1) #check whether renaming worked
```

```python
bpb = bpb.sort_values('Titel') #sort "Titel" column alphabetically
bpb = bpb.reset_index(drop=True) #reset index
bpb.head(5) #check wether sorting and reindexing worked
```

```python
bpb = bpb[~bpb['Link'].str.contains(('73410'), na=False)] #delete useless data 
bpb = bpb[~bpb['Link'].str.contains(('237913'), na=False)] 
bpb = bpb[~bpb['Link'].str.contains(('161896'), na=False)] 
```

```python
bpb.shape #check wether rows where deleted
```

```python
bpb.drop(bpb.tail(12).index,inplace=True) # drop last 12 rows (dead links)
```

```python
bpb.to_csv('bpb_csv.csv', index=False) #save dataframe to csv
bpb.to_excel("bpb_excel.xlsx", index=False)#save dataframe to excel
```
